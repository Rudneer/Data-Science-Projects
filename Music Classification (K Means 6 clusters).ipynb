```python
# Importing all the necessary libraries

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading data from 'music.csv' file

df = pd.read_csv('music.csv')
df
```

```python
plt.scatter(df.age,df.gender)
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['age']])
df['age'] = scaler.transform(df[['age']])
df
```

```python
km = KMeans(n_clusters=6)
y_predicted = km.fit_predict(df[['age','gender']])
y_predicted
```

```python
df['clusters'] = y_predicted
df
```

```python
df1 = df[df.clusters==0]
df2 = df[df.clusters==1]
df3 = df[df.clusters==2]
df4 = df[df.clusters==3]
df5 = df[df.clusters==4]
df6 = df[df.clusters==5]

plt.scatter(df1.age,df1.gender,color='red')
plt.scatter(df2.age,df2.gender,color='orange')
plt.scatter(df3.age,df3.gender,color='yellow')
plt.scatter(df4.age,df4.gender,color='green')
plt.scatter(df5.age,df5.gender,color='blue')
plt.scatter(df6.age,df6.gender,color='black')
# plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')

plt.xlabel('Age')
plt.ylabel('Gender')
plt.legend()
```
